## block from Feb 3 noon to Feb 10 noon where TSP UNCONDITIONAL is ON

```python
wego['DATE'] = pd.to_datetime(wego['DATE'])
```

```python
tsp_half_chunk = wego.loc[wego['DATE'] >= '2025-2-4']
tsp_one_chunk = tsp_half_chunk.loc[tsp_half_chunk['DATE'] <='2025-2-9']
tsp_one_first_day=wego.loc[wego['DATE']=='2025-2-3']
tsp_one_time = tsp_one_first_day.loc[tsp_one_first_day['ACTUAL_ARRIVAL_TIME']>'12:00:00']
tsp_one_last_day=wego.loc[wego['DATE']=='2025-2-10']
tsp_last_time = tsp_one_last_day.loc[tsp_one_last_day['ACTUAL_ARRIVAL_TIME']<'12:00:00']
tsp_one_complete = pd.concat([tsp_one_first_day, tsp_one_chunk,tsp_one_last_day])
tsp_one_complete
```

## (Block from Feb 10 noon to April 28 noon where TSP Schedule-Conditional Priority is ON)

```python
tscpb_chunk = wego[wego['DATE'].between('2025/2/11','2025/4/27')]
tscpb_first_day=wego[wego['DATE']=='2025/2/10']
tscpb_first_time = tscpb_first_day[tscpb_first_day['ACTUAL_ARRIVAL_TIME']>'12:00:00']
tscpb_last_day=wego[wego['DATE']=='2025/4/28']
tscpb_last_time = tscpb_last_day[tscpb_last_day['ACTUAL_ARRIVAL_TIME']<'12:00:00']
tscpb_complete = pd.concat([tscpb_first_time, tscpb_chunk, tscpb_last_time])
tscpb_complete
```

## Block from April 28 noon to May 5 noon where TSP is OFF

```python
tsp_off_chunk = wego[wego['DATE'].between('2025/4/29','2025/5/4')]
tsp_off_first_day = wego[wego['DATE']=='2025/4/28']
tsp_off_first_time = tsp_off_first_day[tsp_off_first_day['ACTUAL_ARRIVAL_TIME']>'12:00:00']
tsp_off_last_day = wego[wego['DATE']=='2025/5/5']
tsp_off_last_time = tsp_off_last_day[tsp_off_last_day['ACTUAL_ARRIVAL_TIME']<'12:00:00']
tsp_off_complete = pd.concat([tsp_off_first_time,tsp_off_chunk,tsp_off_last_time])
tsp_off_complete
```

## (Block from May 5 at noon to May 12 noon where TSP UNCONDITIONAL is ON)

```python
tsp_two_chunk = wego.loc[wego['DATE'].between('2025/5/6','2025/5/11')]
tsp_two_first_day = wego.loc[wego['DATE']=='2025/5/5']
tsp_two_first_time = tsp_two_first_day.loc[tsp_two_first_day['ACTUAL_ARRIVAL_TIME']>'12:00:00']
tsp_two_last_day = wego.loc[wego['DATE']=='2025/5/12']
tsp_two_last_time = tsp_two_last_day.loc[tsp_two_last_day['ACTUAL_ARRIVAL_TIME']<'12:00:00']
tsp_two_complete = pd.concat ([tsp_two_first_time, tsp_two_chunk, tsp_two_last_time])
tsp_two_complete
```
